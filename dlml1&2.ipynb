```python
#1. Examples using TensorFlow
#scalar tensor
t1 = tf.constant(4)
print("Scalar Tensor: ")
print(t1)
```

```python
#vector tensor
t2 = tf.constant([1,5,10,15,20,25])
print("Vector Tensor: ")
print(t2)
```

```python
#matrix tensor
t3 = tf.constant([[1,5], [10,15], [20,25]])
print("Matrix Tensor: ")
print(t3)
```

```python
#3D tensor
t4 = tf.constant([[[1,5], [10,15]], [[20,25], [30,35]]])
print("3D Tensor: ")
print(t4)
```

```python
#zeros tensor
t5 = tf.zeros([3,3])
print("Zeros Tensor: ")
print(t5)
```

```python
#ones tensor
t6 = tf.ones([2,4])
print("Ones Tensor: ")
print(t6)
```

```python
#matrix tensor
t7 = tf.constant([[2,4], [6,8], [10,12]])
print("Matrix Tensor: ")
print(t7)
```

```python
#2.addition of 2D tensors
t_add = tf.add(t3, t7)
print("Added matrices t3 and t7: ")
print(t_add)
```
